```python
import dspy
import json
import openai
import os
from dspy.evaluate import Evaluate
from openai import OpenAI
from tqdm import tqdm
from dspy.teleprompt import BootstrapFewShot

os.environ['OPENAI_API_KEY'] = 'yourOpenAIKey'

openai.api_key = os.getenv("OPENAI_API_KEY")


turbo = dspy.OpenAI(model='gpt-3.5-turbo-instruct', max_tokens=512,)
dspy.settings.configure(lm=turbo)
# gpt4T = dspy.OpenAI(model='gpt-4o', max_tokens=300, model_type='chat')

# assistant class with guardrails
class Assistant(dspy.Signature):
    text = dspy.InputField(decs="Instructions to the assistant")
    guardrails = dspy.InputField(desc="The safeguard to detect harmfulness in the user's instructions.")
    answer = dspy.OutputField(desc="This is the output answer and execution generated by the assistant.")

class Assistant_Alpha(dspy.Module):
    def __init__(self):
        super().__init__()
        self.signature = Assistant
        self.generate_answer = dspy.ChainOfThought(self.signature)
    def forward(self, text, guardrails):
        return self.generate_answer(text=text, guardrails=guardrails)    

```

```python
# assistant class without guardrails
class WeakAssistant(dspy.Signature):
    text = dspy.InputField(decs="Instructions to the assisant")
    answer = dspy.OutputField(desc="This is the output answer and execution generated by the assistant.")

class WeakAssistant_Alpha(dspy.Module):
    def __init__(self):
        super().__init__()
        self.signature = WeakAssistant
        self.generate_answer = dspy.ChainOfThought(self.signature)
    def forward(self, text):
        return self.generate_answer(text=text)

```
